# 过拟合和欠拟合

这个其实很好说明，就放两张图来演示：

神经网络的拟合能力是很强大的，万能近似定理表明：一个前馈神经网络(也就是全连接神经网络)如果具有线性输出层和至少一层具有任何一种“挤压” 性质的激活函数（如sigmoid激活函数）的隐藏层，只要给予网络足够数量的隐藏单元，它可以以任意的精度来近似任何从一个有限维空间到另一个有限维空间的Borel可测函数。

关于什么是Borel可测函数我们不深究，我们现在只要知道一切连续函数都是波莱尔函数即可。

从上面的文字我们可以看出：神经网络很厉害，但是想要厉害起来也需要足够的隐藏单元，也就是参数量。这是神经网络学习的基础，如果隐藏单元少了，那么网络就会出现“学不会、学不精”的现象

![](pic\1.png)

而相反，如果参数量多了的话，神经网络就会钻牛角尖，从理解变成了死记硬背。把错误的情况也记住了。

![](pic\2.png)



```python
import tensorflow as tf
import tensorflow.keras as keras
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data_size = 50

x = np.linspace(-5, +5, data_size, dtype=np.float32) # tensorflow2 默认接受float32数据
noise = np.random.normal(scale=20, size=data_size)

a = 1.3
b = 2
c = 1

y = a * (x ** 3) + b * x + c + noise

plt.scatter(x, y)
```

```python
Input = keras.Input(shape=(1)) # 自变量是个标量

FC1_1 = keras.layers.Dense(10, activation='sigmoid') (Input)
result = keras.layers.Dense(1) (FC1_1) # 线性输出层

Model = keras.Model(inputs=Input, outputs=result)

Model.compile(loss='mse')
Model.fit(x=x, y=y, epochs=5000)
```

```python
pred1 = Model(np.reshape(x, [-1, 1]))

plt.scatter(x, y)
plt.plot(x, pred1, color='r') 
```

```python
FC2_1 = keras.layers.Dense(100, activation='sigmoid') (Input)
FC2_2 = keras.layers.Dense(120, activation='sigmoid') (FC2_1)
FC2_3 = keras.layers.Dense(140, activation='sigmoid') (FC2_2)
FC2_4 = keras.layers.Dense(160, activation='sigmoid') (FC2_3)
FC2_5 = keras.layers.Dense(180, activation='sigmoid') (FC2_4)
FC2_6 = keras.layers.Dense(200, activation='sigmoid') (FC2_5)
result2 = keras.layers.Dense(1) (FC2_6) # 线性输出层

Model2 = keras.Model(inputs=Input, outputs=result2)

Model2.compile(loss='mse')
Model2.fit(x=x, y=y, epochs=5000)
```

```python
pred2 = Model2(np.reshape(x, [-1, 1]))

plt.scatter(x, y)
plt.plot(x, pred2, color='r') 
```

```python
FC3_1 = keras.layers.Dense(24, activation='sigmoid') (Input)
FC3_2 = keras.layers.Dense(32, activation='sigmoid') (FC3_1)
result3 = keras.layers.Dense(1) (FC3_2) # 线性输出层

Model3 = keras.Model(inputs=Input, outputs=result3)

Model3.compile(loss='mse')
Model3.fit(x=x, y=y, epochs=5000)
```

```python
pred3 = Model3(np.reshape(x, [-1, 1]))

plt.scatter(x, y)
plt.plot(x, pred3, color='r')
```

如何判断网络是否过拟合或欠拟合呢？

我们一般训练网络时，通常会把数据独立的划分为2到3份。

以2份为例，这两份数据中，用于训练的我们称为训练集，而另一部分则叫做测试集。

如果网络欠拟合，那么网络通常在训练集和测试集上的表现应该都不高。

若网络拟合的正好，那么通常在训练集上的表现应该不错，并且在测试集上的表现和训练集差不多。就像我们上节课卷积在训练集上99+%，在测试集上98+%

如果网络过拟合，则通常在训练集上会表现得很好，却在测试集上表现得比较差。

一般，欠拟合解决的方法比较单一，就是增加模型的复杂程度。比如单纯的增加网络层，或者使用一些更优秀的网络结构。

而解决过拟合的方法，一般原理都是降低模型的复杂程度，使模型的表达能力下降，即拟合能力下降。

一般常用的方法有减少网络层数(直接减少参数量)，L1\L2正则化，Dropout等等。

其中，正则化是通过限制网络参数的大小来限制网络的表达能力(一般来说参数越大的话，得到的激活值越高，激活函数对数据的挤压扭曲作用越大，而神经网络的表达能力又是和非线性能力相关)

Dropout则是每次训练时，随机丢弃掉一些神经元，这样可以减少神经元之间的依赖性，加强网络的泛化能力。（更多解释：https://www.cnblogs.com/shenxiaolin/p/9841824.html

keras为网络添加正则化或Dropout还是很简单省事的：

首先来看正则化，keras的正则化是以层为单位的，也就是说，我们可以选择我们要限制表达能力的网络层，如果要为整个模型加上正则化，即为对模型中的所有(带有可训练参数的)层加入正则化。

# 梯度消失和梯度爆炸

我们在之前学到了，我们神经网络的学习，运用了梯度下降算法，学习的实质就是讲网络的参数朝着其梯度的反方向进行更新。又由于多层神经网络可以看作是多个函数的复合，而复合函数的求梯度会用到链式法则。

再来看看链式法则：若u = f(x)，y = g(u)则

$$\frac{dy}{dx} = \frac{dy}{du}\cdot\frac{du}{dx}$$

如果复合的次数越多，则链式法则中连乘的次数也越多。我们考虑一个比较大的数n：

 - 1.当网络每一层的单独求偏导，且偏导总体而言小于1时，那么累乘起来将会得到一个很小的数字。这就是梯度消失现象。此时有前面隐藏层的学习速率低于后面隐藏层的学习速率，即随着隐藏层数目的增加，分类准确率反而下降了。
 - 2.相反，如果偏导总体大于1，则累成起来会得到一个很大的数字，这就是梯度爆炸。此时有前面隐藏层的学习速率高于后面隐藏层的学习速率，即随着隐藏层数目的增加，分类准确率反而下降了。

```python
print(0.9 ** 100)
print(0.25 ** 10)
print(1.1 ** 100)
```

sigmoid激活函数是一个比较奇妙的激活函数，若即为S(x)，则

$$\frac{dS}{dx} = S(x)(1-S(x))$$

又因为其取值在零一之间，我们用均值不等式的知识或者直接求极值可以得到，当$S(x) = (1-S(x))$，也就是$S(x)=0.5$时，导数值取最大为$0.25$。正好是$x=0$的时候。

根据上面提的，很明显，当网络层数很深的时候，会出现梯度消失的现象。

relu的提出就解决了sigmoid函数的这个缺点。

这就是之前为什么模型使用relu总是比使用sigmoid要训练的快的原因了。

总结一下：
 - 梯度消失：（1）隐藏层的层数过多；（2）采用了不合适的激活函数(更容易产生梯度消失，但是也有可能产生梯度爆炸)
 - 梯度爆炸：（1）隐藏层的层数过多；（2）权重的初始化值过大
 
解决这两个问题不仅可以换用适合的激活函数，还可以使用一些技巧：如：
